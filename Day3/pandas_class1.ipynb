```python
import pandas as pd
```

```python
A=pd.Series()
print(A)     #note:-its create enmpty file
```

```python
#create series with single element
B=pd.Series("Hello")
print(B)
```

```python
#create series using list
Colors=["blue","green","yelllow","red"]
series=pd.Series(Colors)
print(series)          #note:-[] square brackets using list
```

```python
#create series using tuple
tuple=(1000,2000,3000,4000)
series=pd.Series(tuple)
print(series)            #note:-see the output 0123 these are all index numbers by deafault
                        #note:-tuple we are using parenthesis()
```

```python
#create series using dictionary
dict={101:"vishwa",102:"ram",103:"ravi",104:"basu"}
series=pd.Series(dict)
print(series)
                                #note:-here we give index values so it will print 101 102 like that
                                #note:-in dictionary we used curly braces{}
```

```python
A=pd.Series(["monday","tuesday","wednesday","thursday"])
print(A)    #note:-its default take index values
```

```python
#i want to change index values day1 and day2 like that
A.index=["day1","day2","day3","day4"]
print(A)
```

```python
#create dataframe using nested list
list=[[1,2,3],[4,5,6],[7,8,9]]
A=pd.DataFrame(list)
A
```

```python
data=[["vishwa",40],["ram",50],["ravi",60],["basu",20]]
A=pd.DataFrame(data,columns=["Name","Age"])
A                                                 #note:-we use  columns=["name","age"] get the name and age
```

```python
#create dataframe using dictionary
dict={"Name":["vishwa","ram","ravi","chetan","laxman"],"Age":[45,34,23,45,35],"Job":["HR",",Manager","Analyst","Teacher","Engineer"],"Salary":
      [67000,84000,45000,53000,70000]}
data=pd.DataFrame(dict)
data
```

```python
list=[["apple",40],["mango",60,120],["orange",50]]
A=pd.DataFrame(list,columns=["fruits","no_of_fruits","price_per_kilo"])
A             #note:-see the output it will show the nan/null/none all three says missing values
```
